```python
import numpy as np
import pandas as pd 
```

```python
def convertToInt(x):
    if type(100) == type(x):
        return x
    return np.nan
```

```python
df = pd.read_excel("User/CropYield/crop yield data sheet.xlsx" , converters={
    'Temperatue' : convertToInt,
})
```

```python
for  col in df.columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , random_state=2)
```

```python
y_test
```

```python
scaler = MinMaxScaler()
```

```python
X_tnf_train = scaler.fit_transform(X_train) # type: ignore
X_tnf_test = scaler.transform(X_test) # type: ignore
```

```python
X_tnf_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_tnf_train, y_train)
```

```python
pred = lr.predict(X_tnf_test)
```

```python
y_test.shape
```

```python
pred.shape
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , pred)
```
